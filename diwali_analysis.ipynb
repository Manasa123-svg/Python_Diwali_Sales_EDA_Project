```python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import th the cse file

data=pd.read_csv(r'c:\Users\nimma\Downloads\Diwali Sales Data.csv',encoding_errors='ignore')
```

```python
# information about the dataset

data.info()
```

```python
# remove the unrelated\blank columns

data.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
data.head()
```

```python
# find the null values

data.isnull().sum()
```

```python
# fill the null values(int,float values)

amount_mean=data['Amount'].mean().round(0)

data.loc[:,'Amount']=data['Amount'].fillna(amount_mean)
data.head(20)
```

```python
data.isnull().sum()
```

```python
# change data type float into int

data['Amount']=data['Amount'].astype('int')
```

```python
data['Amount'].dtype
```

```python
# rename the column name

data.rename(columns={'cust_name':'Customer_name'},inplace=True)
```

```python
# describe method returns describe of the dataframe(ie,sum.count,min,max etc)

data.describe()

```

```python
# use describe() for the specific columns

data[['Age','Orders','Amount']].describe()
```

Exploratory Data Analysis

```python
# ploting a bar chart for gender it's count

a=sns.countplot(data=data,x='Gender',hue='Gender')

for bar in a.containers:
    a.bar_label(bar)
```

```python
# plotting a bar chart for gender vs total amount

sales_gen=data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(data=sales_gen,x='Gender',y='Amount',hue='Gender')
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

Age

```python
# plotting a bar chart for age count

ax=sns.countplot(data=data,x='Age Group',hue='Gender')

for bar in ax.containers:
    ax.bar_label(bar)
```

```python
# plotting a bar chart for age grouo vs total amount

sales_age_group=data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(data=sales_age_group,x='Age Group',y='Amount')
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

state

```python
# total number of orders form the top 10 states

sales_state=data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(data=sales_state,x='State',y='Orders',hue='State')
```

```python
# total amount\sales in top 10 states


sales_state=data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})

sns.barplot(data=sales_state,x='State',y='Amount',hue='State')
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

Marital status

```python
# plotting for marital status count

a=sns.countplot(data=data,x='Marital_Status')

for bar in a.containers:
    a.bar_label(bar)
```

 above graphs 0 is a un_married 1 is a married

```python
# plotting bar  marital status and gender vs amount

sales_marital_status_gender=data.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_marital_status_gender,x='Marital_Status',y='Amount',hue='Gender')
```

From above graphs we can see that most of the buyers are un_married (women) and they have high purchasing power

occupation

```python
# plotting bar charts for occupation it's count

sns.set(rc={'figure.figsize':(20,5)})
a=sns.countplot(data=data,x='Occupation',hue='Occupation')

for bar in a.containers:
    a.bar_label(bar)
```

```python
# plotting bar chars for occupations vs amount

amount_occupation=data.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=amount_occupation,x='Occupation',y='Amount',hue='Occupation')
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

product category

```python
# plotting bar chars for product category it's count
sns.set(rc={'figure.figsize':(30,5)})

a=sns.countplot(data=data,x='Product_Category',hue='Product_Category')

for bar in a.containers:
    a.bar_label(bar)
```

```python
# plotting bar chars for occupations vs amount

amount_occupation=data.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(data=amount_occupation,x='Product_Category',y='Amount',hue='Product_Category')
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category

product id

```python
# plotting bar charts form the product id vs orders top 10

sales_product_id = data.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_product_id, x = 'Product_ID',y= 'Orders',hue='Product_ID')
```

Un_Married women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category
